```python
import numpy as nump
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd
import csv


url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.csv'
df = pd.read_csv(url,sep=',')
df.head()
df.to_csv('output.csv')


# assumptions of the perceptron model
# assumes target feature is the last column
# assumes target data has already been classified as numeric values
# assumes only two different classifications

dataframe = pd.read_csv('output.csv', header=None)


def perceptronAlgo(dataf):
  uvalue = 0
  print("Dataframe Shape: ",dataf.shape)
  # Retrieving the number of rows in dataset
  no_rows = len(dataf.index)
  # Getting the number of columns in dataset
  no_features = len(dataf.columns)
  # Generating matrices
  wght_mtrx = nump.zeros(shape=(1,no_features))
  # Extracting features from dataset which helped us in classification
  column_feature_data = dataf.iloc[:,:no_features-1]
  print("Features are: ",column_feature_data.shape)
  samplebias = nump.ones((len(dataf.index),1))
  print(samplebias.shape)
  print(column_feature_data.shape)
  column_feature_data = column_feature_data.to_numpy()
  X = nump.append(samplebias, column_feature_data,1)
  print("Weight Matrix :", wght_mtrx.shape)
  # Getting the expected output vector
  yv = dataf.iloc[:,no_features-1].to_numpy()
  print("Given True y: ",yv.shape)
  i = 0
  while i < no_rows:
    print("Initial weight matrix:",wght_mtrx)
    tempweight = nump.dot(X[i],wght_mtrx.transpose())
    print("Temp weight:",tempweight)
    print(tempweight, yv[i])
    if tempweight < 0:
      tempweight = -1
    if tempweight > 0:
      tempweight = 1
    
    print("Updated temp weight",tempweight, yv[i])
    if tempweight == yv[i]:
      wght_mtrx = wght_mtrx
    else:
      print("Updating")
      uvalue = uvalue +1
      dot_product = nump.multiply(yv[i],X[i])
      print("yv[i],X[i]:",yv[i],X[i])
      print("wght_mtrx += dot_product:",wght_mtrx,dot_product)
      wght_mtrx = dot_product + wght_mtrx
      i+=1
      
    print("-----weight-----",wght_mtrx)
    print("Updating counter value",uvalue)

    print(wght_mtrx)

print(perceptronAlgo(dataframe))
```
